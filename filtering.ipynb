# How do I filter rows of a pandas DataFrame by column value?

```python
import numpy as np
import pandas as pd
```

```python
movies = pd.read_csv('http://bit.ly/imdbratings')
movies.head()
```

```python
movies.shape
```

```python
booleans=[]# tell us which rows match a given condition
for length in movies.duration:
    if length>=200:
        booleans.append(True)
    else:
        booleans.append(False)
```

```python
booleans[0:5]
```

```python
is_long=pd.Series(booleans)
is_long
```

```python
movies[is_long]
```

```python
movies[movies.duration>=200] #same thing as above
```

```python
is_long=movies.duration>=200 #you dont need a for loop in this subject.So how can write this code more simply
is_long.head()
```

```python
movies[is_long]
```

```python
movies[movies.duration>=200].genre
```

```python
movies.loc[movies.duration>=200,'genre'] #we can use loc for filtering
```

## How do I apply multiple filter criteria to a pandas DaraFrame

```python
movies[(movies.duration>=200) & (movies.genre=='Drama')] # if you use and operator instead of (&) raise an error 
```

```python
movies[(movies.genre=='Drama') |(movies.genre=='Action') | (movies.genre=="Crime")]
```

```python
movies[movies.genre.isin(['Crime','Drama','Action'])] #is a trick for simplfy the above code
```

```python
movies[movies.title.str.len()>5] #filter rows use the length
```

```python
movies[movies.title.str.contains('all|love')] #contains function is similar to like statements
```

# Lambda Method for Filtering

# filter()


### Parameters

### like : str
Keep labels from axis for which “like in label == True”.

### regex :  str (regular expression)
Keep labels from axis for which re.search(regex, label) == True.

### axis : {0 or ‘index’, 1 or ‘columns’, None}, default None
The axis to filter on, expressed either as an index (int) or axis name (str). By default this is the info axis, ‘index’ for Series, ‘columns’ for DataFrame.

```python
df6 = pd.DataFrame(np.array(([1, 2, 3], [4, 5, 6])),
                  index=['mouse', 'rabbit'],
                  columns=['one', 'two', 'three'])
df6
```

```python
df6.filter(items=['one','three']) # select column by name
```

```python
df6.filter(regex='e$',axis=1) # at the end of column with 'e' in the columns (axis=1)
```

```python
df6.filter(regex='o$',axis=0) # at the end of column with 'o' in the rows (axis=0)
```

```python
df6.filter(like='bbi', axis=0) # select rows containing 'bbi'
```

```python
df7= pd.DataFrame({'A' : ['foo', 'bar', 'foo', 'bar',
                          'foo', 'bar'],
                   'B' : [1, 2, 3, 4, 5, 6],
                   'C' : [2.0, 5., 8., 1., 2., 9.]})
df7
```

```python
df7.groupby('A').B.min()
```

```python
df7.groupby('A').filter(lambda x : x['B'].min()>1)
```
